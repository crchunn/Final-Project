```python
from sklearn import tree
import pandas as pd
import os
```

```python
music_df = pd.read_csv(os.path.join("dataA.csv"))
music_df.head()
```

```python
del music_df['Unnamed: 0']
music_df
```

```python
target = music_df["mode"]
target_names = ["minor", "major"]
```

```python
data = music_df.drop("mode", axis=1)
feature_names = data.columns
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=33)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
sorted(zip(rf.feature_importances_, feature_names), reverse=True) 
```
