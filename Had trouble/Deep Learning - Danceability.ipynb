```python
# Dependencies
import numpy as np
import pandas as pd
```

```python
import tensorflow
tensorflow.keras.__version__
```

```python
music_df = pd.read_csv('dataA.csv')
music_df.head()
```

```python
del music_df['Unnamed: 0']
music_df
```

```python
year_df = music_df
year_df
```

```python
# Establish the bins
bins = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
group_names = ["2000", "2001", "2002", "2003", "2004", "2005", "2006", "2007", "2008", "2009", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020"]
```

```python
# Categorize year based on the bins
year_df["Music Year"] = pd.cut(year_df["Year"], bins, labels=group_names, include_lowest=True)
year_df
```

```python
X = music_df.drop("danceability", axis=1)
y = music_df["danceability"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
# Step 2: Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Create model and add layers
model = Sequential()
model.add(Dense(units=30, activation='relu', input_dim=14))
model.add(Dense(units=30, activation='relu'))
model.add(Dense(units=2, activation='softmax'))
```

```python
# Compile and fit the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=60,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
encoded_predictions = model.predict_classes(X_test_scaled[:10])
prediction_labels = label_encoder.inverse_transform(encoded_predictions)
```

```python
print(f"Predicted classes: {prediction_labels}")
print(f"Actual Labels: {list(y_test[:10])}")
```
